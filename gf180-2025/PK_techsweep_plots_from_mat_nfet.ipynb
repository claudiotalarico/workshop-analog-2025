```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
from pygmid import Lookup as lk
```

```python
choice = 0  #set to 0 for nmos and 1 for pmos
devices = ['nfet_03v3', 'pfet_03v3']
fet = lk('./'+devices[choice]+'.mat')
```

```python
# Reference Current Density
# weak inversion
K_wi = np.array([ 28, 27 ]) # S/A
# strong inversion
K_si = np.array([10*np.sqrt(2e-6) , 6*np.sqrt(1.7e-6) ]) # S/A/sqrt(A)
# reference current
I0 = (K_si/K_wi)**2
```

```python
# sweep variable vectors
w = fet['W']
l = fet['L']
vgs = fet['VGS']
vds = fet['VDS']
vsb = fet['VSB']
```

```python
print(f'Transistor Width is {w} um and Lengths are {l} um')
```

```python
VDS1=1.65; VSB1=0
```

## Gm/ID versus ID/(W/L)

```python
# plot gm/ID versus ID/(W/L)

fig, ax1 = plt.subplots()
ax1.grid(which='both')
ax1.set_xlabel('$I_{DS}/(W/L)$ (A)')
color = 'tab:blue'
ax1.set_ylabel('$g_m/I_D$ (S/A)', color=color)
# get data
gm_ids = fet.lookup('GM_ID', L=l, VDS=VDS1, VSB=VSB1)
ids = fet.lookup('ID', L=l, VDS=VDS1, VSB=VSB1)
ax1.loglog((ids / w * l[:, None]).T, gm_ids.T)

# tangents
# strong inversion
x = (ids / w * l[:, None]).T[:,1]
y = K_si[choice]/np.sqrt(x) 
ax1.loglog(x,y,'.',color='k')
# weak inversion
y = K_wi[choice] *np.ones(len(x))
ax1.loglog(x,y,'.',color='tab:gray')
# calculate I0
ax1.set_ylim(1, 30)
ax1.legend(loc='center left', title='L', labels=l.tolist())
# fig.tight_layout()
plt.title(devices[choice]+', $V_{DS}$='+str(VDS1)+', $V_{SB}$='+str(VSB1))
label = f"$I_0$={I0[choice]/1e-6:.2f}$\\mu$A"
ax1.text(I0[choice]/3, 3, label, color='m' )
plt.axvline(x=I0[choice], color='m', linestyle='--')
plt.show()
fig.tight_layout()
```

## Gm/ID versus Current Density

```python
# plot gm/ID versus IC = ID/(W/L)/I0
# using I0 from previous cell

fig, ax1 = plt.subplots()
ax1.grid(which='both')
ax1.set_xlabel('$IC$')
color = 'tab:blue'
ax1.set_ylabel('$g_m/I_D$ (S/A)', color=color)

gm_ids = fet.lookup('GM_ID', L=l, VDS=VDS1, VSB=VSB1)
fts = fet.lookup('GM_CGG', L=l, VDS=VDS1, VSB=VSB1)
ids = fet.lookup('ID', L=l, VDS=VDS1, VSB=VSB1)
ax1.loglog((ids / I0[choice] / w * l[:, None]).T, gm_ids.T)

# tangents
# strong inversion
x = (ids / I0[choice] / w * l[:, None]).T[:,1]
K_si_0 = K_si[choice]/np.sqrt(I0[choice]) 
y = K_si_0/np.sqrt(x) 
ax1.loglog(x,y,'.',color='k')
# weak inversion
y = K_wi[choice] *np.ones(len(x))
ax1.loglog(x,y,'.',color='tab:gray')
plt.axvline(x=1, color='m', linestyle='--')

ax1.set_ylim(1, 30)
ax1.legend(loc='center left', title='L', labels=l.tolist())
plt.title(devices[choice]+', $V_{DS}$='+str(VDS1)+', $V_{SB}$='+str(VSB1))

fig.tight_layout()
```

## Gm/ID and fT versus ID/(W/L)

```python
# plot gm/ID & fT versus ID/(W/L)

fig, ax1 = plt.subplots()
ax1.grid(which='both')
ax1.set_xlabel('$I_{DS}/(W/L)$ (A)')
color = 'tab:blue'
ax1.set_ylabel('$g_m/I_D$ (S/A)', color=color)
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('$f_T$ (GHz)', color=color)

ax2.tick_params(axis='y', labelcolor=color)
ax2.grid(True)
# get data
gm_ids = fet.lookup('GM_ID', L=l, VDS=VDS1, VSB=VSB1)
ids = fet.lookup('ID', L=l, VDS=VDS1, VSB=VSB1)
ax1.loglog((ids / w * l[:, None]).T, gm_ids.T)
# tangents
# strong inversion
x = (ids / w * l[:, None]).T[:,1]
y = K_si[choice]/np.sqrt(x) 
ax1.loglog(x,y,'.',color='k')
# weak inversion
y = K_wi[choice] *np.ones(len(x))
ax1.loglog(x,y,'.',color='tab:gray')
# calculate I0
ax1.set_ylim(1, 30)
ax1.legend(loc='center left', title='L', labels=l.tolist(), bbox_to_anchor=(1.2,0.5))

fts = fet.lookup('GM_CGG', L=l , VDS=VDS1, VSB=VSB1)/2/np.pi
ax2.loglog((ids / w * l[:, None]).T, fts.T/1e9)

plt.title(devices[choice]+', $V_{DS}$='+str(VDS1)+', $V_{SB}$='+str(VSB1))
label = f"$I_0$={I0[choice]/1e-6:.2f}$\\mu$A"
ax1.text(I0[choice]/3, 3, label, color='m' )
plt.axvline(x=I0[choice], color='m', linestyle='--')
plt.show()
fig.tight_layout()
```
